```python
cardinalBE = cv2.imread("C:/Users/venka/OneDrive/Desktop/Big Data/Training/cardinalBE.png")
cardinalBE=cv2.cvtColor(cardinalBE, cv2.COLOR_BGR2GRAY)
cardinalBE.shape
```

```python
# Dividing the Cardinal image into 16*16 blocks

def BlockVector(img):
    featureV=[]
    for row in range(0,img.shape[0]-16, 16):
        for col in range(0, img.shape[1]-16, 16):
            feature1 = img[row:row+16, col:col+16]
    #         print(feature1)
            x = feature1.flatten()
            featureV.append(x)
    return(featureV)
```

```python
featureV1 = BlockVector(cardinalBE)
cardinalFeatureV = pd.DataFrame(featureV1)
cardinalFeatureV
```

```python
# Adding label as 0 for cardinal feature space.
cardinalFeatureV['label'] = 0*(len(cardinalFeatureV))
cardinalFeatureV
```

```python
# Sliding block vectors

def SlidingVector(img):
    slidingV=[]
    for row in range(0,img.shape[0]-16, 1):
        for col in range(0, img.shape[1]-16, 1):
            feature1 = img[row:row+16, col:col+16]
    #         print(feature1)
            x = feature1.flatten()
            slidingV.append(x)
    return(slidingV)
```

```python
# Code to read all the images in a folder and generate feature vectors.

for filename in os.listdir('C:/Users/venka/OneDrive/Desktop/Big Data/Output images'):
    label=0
    if filename.endswith('.jpg') or filename.endswith('.png'):
        img = cv2.imread(os.path.join('C:/Users/venka/OneDrive/Desktop/Big Data/Output images', filename), cv2.COLOR_BGR2GRAY)
        
        featureV=[]
        for row in range(0,img.shape[0]-16, 16):
            for col in range(0, img.shape[1]-16, 16):
                feature1 = img[row:row+16, col:col+16]
                x = feature1.flatten()
                featureV.append(x)
        df = pd.DataFrame(featureV)
        df['label']=label
        label+=1
        print(f'dataframe of {filename}',df)
        
        
```
